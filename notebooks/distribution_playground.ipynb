```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
def sample_and_plot_beta(alpha_param, beta_param, sample_size=100000):
    """
    Sample values from a Beta distribution and plot the histogram.

    Parameters:
    - alpha (float): The alpha parameter of the Beta distribution.
    - beta (float): The beta parameter of the Beta distribution.
    - sample_size (int): The number of samples to draw.
    """
    # Sample data
    data = beta.rvs(alpha_param, beta_param, size=sample_size)

    # Plot histogram
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, density=True, alpha=0.75, color='blue')
    
    # Plot the theoretical density function
    x = np.linspace(beta.ppf(0.01, alpha_param, beta_param), beta.ppf(0.99, alpha_param, beta_param), 100)
    plt.plot(x, beta.pdf(x, alpha_param, beta_param), 'r-', lw=2, label='Theoretical PDF')
    
    # Add title and labels
    plt.title('Sampled Beta Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

```python
# Example usage
sample_and_plot_beta(alpha_param=2, beta_param=5)
```
