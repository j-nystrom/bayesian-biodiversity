```python
import numpy as np
import matplotlib.pyplot as plt
import arviz as az
```

```python
rng = np.random.default_rng()
```

```python
def plot_distribution(data, distribution):
    
    az.plot_dist(
        data,
        figsize=(6, 3),
        kind="hist",
        color="C1",
        hist_kwargs=dict(alpha=0.6, bins=50),
    )
    az.plot_dist(
        data,
        kind="kde",
        color="C0",
        label="Density",
    )

    plt.title(f"{distribution} distribution")

    plt.tick_params(axis="x", labelsize=10)
    plt.tick_params(axis="y", labelsize=10)

    max_ticks = 15
    ax = plt.gca()
    ax.xaxis.set_major_locator(plt.MaxNLocator(max_ticks))
    plt.xticks(rotation=45)
    plt.show()
```

```python
data = rng.beta(a=1.5, b=3, size=10000)
plot_distribution(data, "Beta")
```

```python
data = rng.gamma(shape=3, scale=2, size=10000)
plot_distribution(data, "Gamma")
```
